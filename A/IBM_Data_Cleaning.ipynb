```python
#dropping the columns that are not required for modelling
df1= df.drop(["Lead_name","Contact_no","POC_name","Lead_POC_email"], axis=1)
```

```python
df1.head()
```

```python
#checking for missing values
df1.isnull().sum()
```

```python
df1[df1["Industry"].isnull()]
```

```python
#replacing the missing values and replacing it with mode
df1["Industry"] = df1["Industry"].fillna((df1["Industry"].mode()[0]))
```

```python
df1[df1["Deal_value"].isnull()]
```

```python
df1["Deal_value"]= df1["Deal_value"].str.replace("$","")
df1.Deal_value.head()
```

```python
#converting Deal_Value to numeric
df1["Deal_value"] = pd.to_numeric(df1["Deal_value"])
#imputing missing value in Deal_value
df1["Deal_value"] = df1["Deal_value"].fillna(df1["Deal_value"].mean())
```

```python
df1[df1["Weighted_amount"].isnull()]
```

```python
# removing the $ sign
df1["Weighted_amount"]= df1["Weighted_amount"].str.replace("$","")
#converting Weighted_amount to numeric
df1["Weighted_amount"] = pd.to_numeric(df1["Weighted_amount"])
df1["Weighted_amount"] = df1["Weighted_amount"].fillna(df1["Weighted_amount"].mean())
```

```python
#converting the Date_of_creation to date time
df1["Date_of_creation"] = pd.to_datetime(df1["Date_of_creation"])
df1.Date_of_creation.dtype
```

```python
df1[df1["Geography"].isnull()].head()
```

```python
df1["Geography"] = df1["Geography"].fillna((df1["Geography"].mode()[0]))
```

```python
df1.isnull().sum()
```

```python
df1[df1["Location"].isnull()]
```

```python
#we see all missing region in usa area,so replacing it with mode of usa area
df2=df1[df1["Geography"]=="USA"]
```

```python
df1["Location"] = df1["Location"].fillna((df2.Location.mode()[0]))
```

```python
df1["Last_lead_update"] = df1["Last_lead_update"].fillna((df1.Last_lead_update.mode()[0]))
```

```python
df1["Resource"] = df1["Resource"].fillna((df1["Resource"].mode()[0]))
```
